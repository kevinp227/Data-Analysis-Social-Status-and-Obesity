## Background & Prior Work

*Obesity is generally defined as an excess amount of body fat as described by the body mass index. An adult who has a BMI of 25-29.9 is overweight and an adult who has a BMI over 30 is obese. A person is morbidly obese (extreme obesity) if their BMI is over 40, however for simplicity any BMI rating over 30 will be categorized as obese. In the United States, as of 2019, one in three adults were considered to be obese (3). Worldwide, it is widely known that lower socioeconomic status is a factor that can lead to health issues. Three main indicators of someone who is of lower socioeconomic status are poverty, unemployment and homelessness. Based on previous research on the topic, adult obesity in the United States is found to be correlated with poverty, unemployment, and homelessness. A report done by the Endocrine Research Unit of Mayo Clinic analyzed how poverty in the United States affected the frequency of diabetes, obesity, and sedentariness. The report concluded that individuals who reside in “poverty-dense counties are those most prone to obesity”, mainly due to less access to affordable healthy food and physical exercise resources (1). Poor families with less money allocated for food usually choose high fat foods that are high in calories, such as cereals, potatoes, processed meat products, etc. that won’t spoil as easily as opposed to fresh fruits and vegetables and lean meats. The same basic concept could be said about physical activity resources for those that are living in poverty. Those with less money have less money to invest in organized activities. The combination of unhealthier food and less physical activity can lead to an individual becoming obese. Another report from the Yale School of Medicine found that “the majority (57%) of chronically homeless adults were overweight or obese”, comapared to the national obesity rate of 42.4 percent. One of the primary reasons for this pattern is the quality of food that the homeless population has access to. Their diet often comes from soup kitchens and shelters, whose food is found to be “low in vitamins and to exceed fat, energy, and protein content recommendations” (4). In the United States, nutritional and healthy food is often expensive and limited in impoverished areas, especially among the homeless population. Another cause for poor health and obesity is restricted access to healthcare and medical services. Due to the cost of professional medical health in the United States, the homeless community generally cannot afford to receive medication, counseling, and treatment regarding their physical health. Furthermore, an article written by Gallup, a wellbeing tracking website, indicates that Americans who have been out of work for more than a year are more likely to become obese than those who have only been unemployed for a short amount of time. Those who have been unemployed for more than a year are 32.7% more likely to become obese while those who have been unemployed for only a few weeks are 22.8% likely to become obese (5). Unemployment may cause some people to engage in behaviors that lead to health problems, while pre-existing health conditions may make it harder for others to find and keep work, perpetuating a negative cycle of worsening health. Much like poverty, unemployment makes it so that people’s financial status is harder for them to access healthy foods.*

- State Unemployment
    - Our data of state unemployment rates is taken from a dataset published by the U.S. Bureau of Labor Statistics, which is a widely used and respected source of information acquired by the United States government.This data is self-reported, and confidentially harvested, leading us to have good faith in the data. Since the mean and the median are close to one another, and the standard deviation is relatively small, there are not any errant outliers or inappropriate data values. The trustworthiness and consistency of this data is high. 
- State Homelessness
    - This data is taken from the HUD Exchange, which works with local governments and Public Housing Authorities to get accurate data on the homeless population for each state. The source of this data is valid and trustworthy. However, upon completing EDA, we found that there is one extreme outlier in this dataset. We concluded that this outlier is reasonable because it measured the magnitude of the homeless population in the state of California, which has a much greater population in comparison to all other states. While we are retaining this outlier, our unemployment dataset maintains good faithfulness.
- State Obesity
    - The dataset measuring obesity rates per state in the United States is taken from data published by the Center for Disease Control and Prevention. This organization is the preeminent source for information regarding health and diseases in the United States, so the data can be assumed to be trustworthy. There is one null observation in this dataset, which is the obesity rate for New Jersey in the year 2018. While the distribution of the data is greatly varied over each state, this makes intuitive sense as it is known that some regions of the United States have much greater obesity rates than average due to regional culture and demographics. There are no unreasonable outliers in this dataset.
- State Poverty 
    - The source of our dataset measuring the population of impoverished people from a sample is taken from the U.S. Census Bureau, a reputable and valid organization that specializes in collecting data from U.S. citizens. Our exploratory analysis shows that this data varies greatly per state, and standard deviation is high. This fact can be attributed to the data being a quantity of impoverished people rather than the rate of impoverished people. Since states such as California and Texas have a high population in proportion to other states, their observations will be much greater than others. # Create Univariate Linear Regression and SSE Function

```python
def univariateLR(X,Y):
    ## Create A, the augmented data array

    ones = np.ones(len(X))

    A1 = np.vstack([ones,X]).T
    #A1.shape

    ## Solve for w, the weight vector

    w1 = np.linalg.lstsq(A1,Y, rcond=None)[0]
    y_pred1 = np.matmul(A1, w1)

    print("Model 1:")
    print("(Obesity) = %f  + %f * (X)" % (w1[0],w1[1]))
    ## Create A, the augmented data array

    ones = np.ones(len(X))
    squares = np.square(X)

    A2 = np.vstack([ones, X, squares]).T

    ## Solve for w, the weight vector
    w2 = np.linalg.lstsq(A2,Y, rcond=None)[0]
    y_pred2 = np.matmul(A2, w2)

    print("Model M2:")
    print("(Obesity) = %f  + %f * (X) + %f * (X)^2" % (w2[0],w2[1],w2[2]))
    ## Create a smooth set of X values for plotting the model
    lineinput = np.linspace(0,10,100)     #replace with how we want to space out the information and the range
    ## Send the X values for plotting through the linear model
    ones = np.ones(len(lineinput))
    squares = np.square(lineinput)
    A3 = np.vstack([ones, lineinput]).T
    A4 = np.vstack([ones,lineinput,squares]).T

    yplot1 = np.matmul(A3, w1)
    yplot2 = np.matmul(A4, w2)
    ## Plot the data along with the model

    # Plot the data
    # Plot the model
    fig, axs = plt.subplots(1, 2)
    plt.subplots_adjust(left=None, bottom=None, right=2, top=None, wspace=None, hspace=None)

    plt.subplot(1,2,1)
    plt.scatter(X, Y, label = "Data")
    plt.plot(lineinput, yplot1, label = "Model 1")
    plt.ylabel(' ') # fill in Y
    plt.xlabel(' ') # fill in X
    plt.title('Model 1')
    plt.legend()

    plt.subplot(1,2,2)
    plt.scatter(X, Y, label = "Data")
    plt.plot(lineinput, yplot2, label = "Model 2")
    plt.ylabel(' ') # fill in Y
    plt.xlabel(' ') # fill in X
    plt.title('Model 2')
    plt.legend()
    plt.show # This lets you plot multiple inputs on the same graph
    model1_SSE = sum((y_pred1 - Y)**2)
    model2_SSE = sum((y_pred2 - Y)**2)

    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    x = ['Model 1', 'Model 2']
    y = [model1_SSE, model2_SSE]
    ax.bar(x,y)
    plt.ylabel('Error')
    plt.title('SSE for Models')
    plt.show()
```

## Create Multivariate Linear Regression Function

```python
def multivar_LR(X1,X2,Y):
    A = np.vstack([X1, X2]).T

    A_train, A_test, Y_train, Y_test = train_test_split(A, Y, test_size=0.20)

    x1 = A_train[:,0]
    x2 = A_train[:,1]


    ones = np.ones(len(A_train))
    A_tr = np.vstack([ones, x1, x2]).T

    W2 = np.linalg.lstsq(A_tr,Y_train, rcond = None)[0]

    Y_train_pred = np.matmul(A_tr,W2)


    A_te = np.vstack([np.ones(len(A_test)), A_test[:,0], A_test[:,1]]).T

    Y_test_pred = np.matmul(A_te,W2)


    SSE_train = sum((Y_train_pred - Y_train)**2)


    SSE_test = sum((Y_test_pred - Y_test)**2)


    print("Obesity = %f  + %f * (X1) + %f * (X2)" % (W2[0], W2[1], W2[2]))


    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    x = ['Training', 'Testing']
    y = [SSE_train, SSE_test]
    ax.bar(x,y)
    plt.ylabel('Error')
    plt.title('SSE of training data and testing data')
 
    plt.show()
```
